```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#below is 1. a)
households = 2000
mean = 220
stdDev = 50  
# generating random data, there was no data provided should be approximately normal distribution 
waterData = np.random.normal(loc=mean, scale=stdDev, size=households)

#creating the x y labels and title
plt.xlabel("Litres(L)")
plt.ylabel("Frequency")
plt.hist(waterData, bins=50, color='black', label='Household Water Use')
plt.title("Total Oakville Residential Water Use since 2005 ")


# drawing mean and standard deviation line's
for i in range(-3, 4):
    plt.axvline(mean + i * stdDev, color='red', linestyle='dashed', linewidth=1, label=f'Mean Â± {i}SD' if i == 1 else "")

plt.legend()
plt.grid(True)
plt.show()

```

```python
# 1. b) prob between 150 and 250 L/day
pBetween = norm.cdf(250, loc=mean, scale=stdDev) - norm.cdf(150, loc=mean, scale=stdDev)
print(f"prob of household using between 150 and 250 L/day is {pBetween:.4f}")
```

```python
# 1. c) prob of using more than 340 L/day
pAbove340 = 1 - norm.cdf(340, loc=mean, scale=stdDev)
print(f"Probability of a household using more than 340 litres/day: {pAbove340:.4f}")
```

```python
# 1. d) prob of using less than 112 L/day
pBelow112 = norm.cdf(112, loc=mean, scale=stdDev)
print(f"Probability of a household using less than 112 litres/day: {pBelow112:.4f}")
```

```python
from scipy.stats import binom, norm

diceRolls = 20          # Num dice rolls
probRoll5 = 1/6         # Prob of rolling five

#3. a) part 1 using a simulation to calcualte prob of rolling the 5, not sure if this is what you meant by simulate.
# Simulate rolling 20 dice
diceRolls = 20
rolls = np.random.randint(1, 7, size=diceRolls)
countFives = np.sum(rolls == 5)

# Calculate the percentage of fives rolled
percentageFives = (countFives / diceRolls) * 100
print(f"fives rolled: {countFives}")
print(f"Percentage of rolling fives: {percentageFives:.2f}% \n\n")
#------------------------------------------------------------------

#3. a) part 2 Prob. of min 7 fives using binomial distribtion
atleast7 = 7
probAtLeast7 = 1 - binom.cdf(atleast7 - 1, diceRolls, probRoll5)
print(f"Probability of minimum 7 fives binom: {probAtLeast7:.6f}")

#3. b) Normal approximation with continuity correction
mean = diceRolls * probRoll5  # Correct mean calculation
#std deviation
sigma = np.sqrt(diceRolls * probRoll5 * (1 - probRoll5))
# Continuity correction
xContinuity = atleast7 - 0.5  
#z score
z = (xContinuity - mean) / sigma
probAtLeast7Normal = 1 - norm.cdf(z)
print(f"\nApproximate probability (normal): {probAtLeast7Normal:.6f}")
```
